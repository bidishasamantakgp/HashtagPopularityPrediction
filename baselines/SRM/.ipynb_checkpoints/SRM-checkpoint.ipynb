```python
import matplotlib.pyplot as plt
import matplotlib.dates as md
import dateutil
import csv
import datetime
import matplotlib.dates as mdates
%matplotlib notebook

from srm.hashtags import hashtag_count_vs_time
from srm.db_manager import connect_to_db, execute_sql
```

#### Establish a DB connection

```python
cursor_mysql, conn = connect_to_db("10.5.23.213", "root", "root")
```

```python
top_hashtags = [ x[0] for x in execute_sql("SELECT hashtag  FROM srm.BigBillionDayHashtags \
                                           group by hashtag order by count(*) desc limit 25")]
#top_hashtags = ('flipkartsale', 'Fraudkart', 'flopkart')
hash_tag_count, hashtag_reply_timestamps = hashtag_count_vs_time("srm", "BigBillionDay", top_hashtags, True)
```

#### This function plots the graph for a single hashtag but in detail. 

```python
hashtag = 'flipkartsale' 
dates = [x[0] for x in hash_tag_count[hashtag]]
plt_data = [x[1] for x in hash_tag_count[hashtag]]
date_markers = [x[0] for x in hashtag_reply_timestamps[hashtag]]
date_markers = list(set(date_markers))
#date_markers.sort()
#Find the positions of replies in the resulting graph axis    
date_indices = []
dates.sort()
for y in date_markers :
    for x in dates :
        #print y, x
        if y<=x :
            date_indices.append(x)
            #print y, x
            break
                
date_indices = list(set(date_indices))
        
plt.subplots_adjust(bottom=0.2)
plt.xticks( rotation=50 )

ax=plt.gca()
#ax.set_xticks(dates)
xfmt = md.DateFormatter('%d %H:%M')
ax.xaxis.set_major_formatter(xfmt)
ax.grid(True)
ax.tick_params(axis='both', which='major', pad=5,  labelcolor="black")
plt.rcParams['xtick.labelsize'] = 10 
plt.plot(dates,plt_data)
plt.xticks(date_markers)
plt.title(hashtag, fontsize=18)
plt.xlabel('Reply timestamp', fontsize=16)
plt.ylabel('Tweets containing the hashtag', fontsize=16)
plt.tight_layout()
#plt.subplot_tool()
plt.show()
print hash_tag_count.keys()
```
