```python
def param_estimate(t_train, T):
    """ For accurate calculation, use the last timestamp from t_train as T and the rest shall be passed as 
    commandline arguments to estimate parameters
    """
    t_args = "\t".join(str(x) for x in t_train[:-1])
    #print command_arg
    res = !./RPP  $T $t_args
    #print res
    l, mu, sigma = (float(x) for x in "".join(res).split('\t'))
    #print l, mu, sigma
    return l, mu, sigma

def predict(l, mu, sigma, T, nd, tp) :
    m = 10 # Prior bilief. Make sure m_m in mic_model.cpp has the same value
    f_tp = norm.cdf((math.log(tp) - mu)/sigma)
    norm.cdf(f_tp)    
    f_T = norm.cdf((math.log(T) - mu)/sigma)
    norm.cdf(f_T)
    cn = (m + nd)*math.exp((f_tp - f_T)*l) - m
    #print "Predicted", int(math.ceil(cn))
    #print cn
    try:
        te = int(math.ceil(cn))
        return te
    except:
        #print cn
        return -1
    
def mape(ta, tp):
    """Calculated MAPE for actual values ta and predicted values tp
    """
    diff = 0
    n = len(ta)
    for i in range(n):
        diff+= abs(ta[i]-tp[i])/float(ta[i])
    return diff/n
```

```python
f = open("newFile.txt")
topic_dict = {}
count = 0
for line in f:
    #print line
    tokens = line.split("\t")
    
    if(len(tokens) < 4):
        continue
    
    #print topicId
    
    try:
        topicId = tokens[3].strip().split(":")[0].split("=")[1]
        #print topicId
        count = count + 1
        tweetId = tokens[0].strip()
        if topicId in topic_dict.keys():
            topic_dict[topicId].append(tweetId)
        else:
            topic_dict[topicId] = [tweetId]
    except:
        continue
print count, len(topic_dict.keys())
#for key in topic_dict.keys():
#    print len(topic_dict[key])
```

```python
timeSeriesDict = {}
prediction_dict = {}
for key in topic_dict.keys():
    original = []
    listParentIds = topic_dict[key]
    for parent_id in listParentIds:
        retweet_times = [row[0] for row in execute_sql("select created_at from srm.BigBillionDay where parent_id_str \
                                                            =%s order by created_at", parent_id)]
        if(len(retweet_times)==0):
            continue
        parent_created_time = [row[0] for row in execute_sql("select created_at from srm.BigBillionDay where id_str \
                                                            =%s", parent_id)]
        created_times = parent_created_time #Stores time at which the tweet and retweets are created
        created_times.extend(retweet_times)
        t_original = list()
        for ts in created_times : 
            t_original.append(int(time.mktime(ts.timetuple())))
        if(len (t_original) <= 10 ):
            continue
        timeSeriesDict[key] = t_original
        original = sorted(original + t_original)
    reply_index = int(len(original)*0.7)
    original_0 = original[0]
    original_n =  original[-1]
    original = [(x - original[0] + 1) for x in original]
    t_train, t_test = original[:reply_index], original[reply_index:]
    
    T = t_train[-1] 
    print T
    #print "time_stamp_dict", timestamp_dict
    
    l, mu, sigma = param_estimate(t_train, T)
    
    #parameter_dict[key] = (l, mu, sigma, nd)

    nd = len(t_train[:-1])
    for tp in t_test:
        ne = predict(l, mu, sigma, T, nd, tp)
        if ne == -1: 
            continue
        if not prediction_dict.has_key(key):
            prediction_dict[key] = [(tp+original_0, original.index(tp)+1, ne)]
        else :
            #print "here"
            prediction_dict[key].append((tp+original_0, original.index(tp)+1, ne))
    
#print "prediction dict", prediction_dict
#print timeSeriesDict

plotMultipleHistograms(original_0, original_n, timeSeriesDict)
    #break
    
    #timestamp_dict[parent_id] = t_original 

        
```

Calculate cumulative MAPE
